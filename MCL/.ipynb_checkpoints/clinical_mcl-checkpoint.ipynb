```python
# create dictionary for cancer_bengin and diagdate 0830 / important. step 1

def max_date(date1, date2):
    if (date1 != date1 or len(str(date1))  == 0) and (date2 != date2 or len(str(date2)) == 0):
        return ''
    elif (date1 != date1 or len(str(date1))  == 0) and (date2 == date2 or len(str(date2)) > 0):
        return date2
    elif (date2 != date2 or len(str(date2))  == 0) and (date1 == date1 or len(str(date1)) > 0):
        return date1
    else:
        return max(date1, date2)

df = pd.read_csv('/nfs/masi/MCL/file/clinical/LabelFile/0421/Report_part1_nodup.csv')
subj_diag = {}
for i, item in df.iterrows():
    # if cancer, use the eariler date, if non-cancer, use the later date
    # if cancer and not cancer, use the cancer date
    if item['MCL_ID'] == item['MCL_ID'] and item['cancer_bengin'] == item['cancer_bengin']:
        mclid = str(item['MCL_ID']).replace('.0', '') 
        if mclid not in subj_diag.keys():
            subj_diag[mclid] = (item['diagdate_norm'], item['cancer_bengin'])
        
        elif subj_diag[mclid][1] == 0 and item['cancer_bengin'] == 1:
            subj_diag[mclid] = (item['diagdate_norm'], item['cancer_bengin'])
        elif subj_diag[mclid][1] == 1 and item['cancer_bengin'] == 0:
            continue
        else:
            subj_diag[mclid] = (max_date(item['diagdate_norm'], subj_diag[mclid][0]), item['cancer_bengin'])
            
            
df = pd.read_csv('/nfs/masi/gaor2/data/MCL/clinical_mcl/combine_update0830.csv')
diagdate, gt_reg = [], []
#with_2yr = []
for i, item in df.iterrows():
    tmp_vec = re.split('[time]', item['id'])
    subj, sess = tmp_vec[0], tmp_vec[-1]
    try:
        assert len(sess) == 8
    except:
        print (item['id'])
    if subj in subj_diag.keys():
        diagdate.append(subj_diag[subj][0])
        gt_reg.append(subj_diag[subj][1])
    else:
        diagdate.append('')
        gt_reg.append('')
df['diagdate'] = diagdate
df['gt_reg'] = gt_reg

df.to_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/mcl/MCL20200830_withconfid.csv', index = False)
```

```python
# get mcl diag_dis 0830 / important.  step 2
from datetime import date
df = pd.read_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/mcl/MCL20200830_withconfid.csv')
subj_sess_dict = {}
for i, item in df.iterrows():
    tmp_vec = re.split('[time]', item['id'])
    subj = tmp_vec[0]
    if item['id'] != item['id']:
        continue
    if subj not in subj_sess_dict.keys():
        subj_sess_dict[subj] = []
    subj_sess_dict[subj].append(item['id'])

diag_dis = []
for i, item in df.iterrows():
    tmp_vec = re.split('[time]', item['id'])
    subj = tmp_vec[0]
    if item['diagdate'] == item['diagdate'] and item['gt_reg'] == 1 :
        last_datestr = str(item['diagdate']).replace('.0', '')
    else:
        last_datestr = max(subj_sess_dict[subj])[-8:]
    sess_datestr = item['id'][-8:]
    last_date = date(int(last_datestr[:4]), int(last_datestr[4:6]), int(last_datestr[6:8]))
    sess_date = date(int(sess_datestr[:4]), int(sess_datestr[4:6]), int(sess_datestr[6:8]))
    delta = last_date - sess_date
    diag_dis.append(delta.days / 365.)
df['diag_dis'] = diag_dis

df.to_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/mcl/MCL20200830_withconfid.csv', index = False)

```

```python
# get diagnosis norm and diag_dis

def convert_norm(diag_str):
    if diag_str != diag_str:
        return ''
    if len(diag_str) == 0:
        return ''
    tmp_vec = re.split('[ -]', str(diag_str))
    if len(tmp_vec) < 3:
        return ''
    norm = tmp_vec[0] + tmp_vec[1] + tmp_vec[2]
    if len(norm) != 8:
        return ''
    return norm

def comput_dis(diag_date, sess_date):
    diag = date(int(diag_date[:4]), int(diag_date[4:6]), int(diag_date[6:8]))
    sess = date(int(sess_date[:4]), int(sess_date[4:6]), int(sess_date[6:8]))
    delta = diag - sess
    return delta.days / 365.

# reference from Jasper
df = pd.read_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/mcl/MCL20200830_withconfid.csv')

sess_diagnorm, sess_diagdis = {}, {}

for i, item in df.iterrows():
    sess_diagnorm[item['id']] = item['diagdate']
    sess_diagdis[item['id']] = item['diag_dis']

df = pd.read_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/mcl/MCL_norm20200830.csv')
diag_norm, diag_dis = [],[]
for i, item in df.iterrows():
    norm = convert_norm(item['Diagnosis.Date.1'])
    if len(norm) > 0:
        diag_norm.append(norm)
    elif item['usedsess'] in sess_diagnorm.keys():
        diag_norm.append(sess_diagnorm[item['usedsess']])
    else:
        diag_norm.append('')
        
    if norm == norm and len(norm) > 0 and 'default' not in item['usedsess']:
        diag_dis.append(comput_dis(norm, item['usedsess'][-8:]))
    elif item['usedsess'] in sess_diagdis.keys():
        diag_dis.append(sess_diagdis[item['usedsess']])
    else:
        diag_dis.append(0)
        
diag_dis = [round(i, 4) for i in diag_dis]
df['diagnorm'] = diag_norm
df['diag_dis'] = diag_dis
df.to_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/mcl/MCL_norm20200830.csv', index = False)
```

```python
# add lastctage to jasper file
def convert_to_time1(string):
    month_dict = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
    str_vec = re.split('[ ,]', string)
    str_vec = [i for i in str_vec if len(i) > 0]
    print (str_vec)
    assert len(str_vec) == 3
    return datetime(int(str_vec[2]), month_dict[str_vec[0]], int(str_vec[1]))

import re
from datetime import datetime
df  = pd.read_csv('/nfs/masi/MCL/file/clinical/LabelFile/0421/Riqiang_9_24_2020_21_15_21.csv')
subj_sess = {}
for i, item in df.iterrows():
    tmp_list = re.split('[_-]', item['XNAT_CTSESSIONDATA ID'])
    subj = tmp_list[0]
    if tmp_list[0] == tmp_list[1]:
        sess = tmp_list[2]
    else:
        sess = tmp_list[1]
    if len(sess) != 8 and len(tmp_list) >=4:
        sess = tmp_list[3] + tmp_list[1] + tmp_list[2]
    try:
        assert len(sess) == 8
    except:
        print (item['XNAT_CTSESSIONDATA ID'])
        continue
    if subj not in subj_sess.keys():
        subj_sess[subj] = []
    subj_sess[subj].append(sess)

df = pd.read_csv('/nfs/masi/MCL/file/clinical/LabelFile/0421/Report_part1_nodup.csv')
lastctage = []
for i, item in df.iterrows():
    mclid = str(item['MCL_ID']).replace('.0', '')
    if mclid in subj_sess and item['Date of Birth'] == item['Date of Birth']:
        lastct = max(subj_sess[mclid])
        age = datetime(int(lastct[:4]), int(lastct[4:6]), int(lastct[6:8])) - convert_to_time1(item['Date of Birth'])
        lastctage.append(round(age.days / 365.0, 2))
    else:
        lastctage.append('')
df['lastctage'] = lastctage
df.to_csv('/nfs/masi/MCL/file/clinical/LabelFile/0421/Report_part1_nodup.csv', index = False)
```

```python
# normalize jasper sub
df = pd.read_csv('/nfs/masi/MCL/file/clinical/LabelFile/0421/subclinical_part1_nodup.csv')
# edu: 1: 'less than hs', 2: 'high school',34. post high school traing, 4: some college, 5. college, 6. post-graduate
edu_dict = { 'Less Than HS': 1, 'High School': 2, 'Post High School Training': 3, 
            'Some College': 4, 'College': 5, 'Post-Graduate': 6}

race_dict = {'African American': 0.5, 'Caucasian': 0, 'Native Hawaiian and Other Pacific Islander': 0.75,
            'Asia': 1, 'Native American': 1}

gender_dict = {'Male': 0, 'Female': 1}

smo_dict = {'Current smoker': 1, 'Ex-smoker': 0.5, 'Never smoked': 0.5}

# see nodule size log, and blood mark log in compare/clinical_spore/mcl_data.ipynb

edu, race, age, fhist, gender, smo, phist = [], [], [], [], [], [], []
bmi = []

for i, item in df.iterrows():
    if item['edu'] in edu_dict.keys():
        edu.append((edu_dict[item['edu']] - 1) / (7 -1) + 0.01)
    else:
        edu.append('')
    
    if item['race'] in race_dict.keys():
        race.append(race_dict[item['race']] + 0.01)
    else:
        race.append('')
        
    if item['gender'] in gender_dict.keys():
        gender.append(gender_dict[item['gender']])
    else:
        gender.append('')

    if item['smo'] in smo_dict.keys():
         smo.append(smo_dict[item['smo']] + 0.01)
    else:
        smo.append('')
        
    if item['age'] == item['age']:
        age.append(( item['age'] - 54) / (80 - 54) + 0.01)
    else:
        age.append('')
        
    if item['bmi'] == item['bmi']:
        bmi.append(( min(60, max(20, item['bmi'] )) - 20) / (60 - 20) + 0.01)
    else:
        bmi.append('')
        
    
        

df['norm_age'] = age
df['norm_bmi'] = bmi       
df['norm_edu'] = edu
df['norm_race'] = race
df['norm_gender'] = gender
df['norm_smo'] = smo

df.to_csv('/nfs/masi/MCL/file/clinical/LabelFile/0421/subclinical_part1_nodup.csv', index = False)


```

```python
# copy from compare/spie21/data.ipynb

df = pd.read_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/spore/spore_with1yr_byscan.csv')
#df = df.query('noempty == 1')
#need_factor = ['age','education','bmi','copd','phist','fhist','smo_status','quit_time','race', 'pkyr']
age = df['age'].tolist()
norm_age = [(i - 54) / (80 - 54) + 0.01 for i in age]
# age: 54, 80

education = df['edu'].tolist()

norm_education = [(i - 1) / (7 -1) + 0.01 for i in education] 
# edu: 1, 7
# 

bmi = df['bmi'].tolist()
bmi = [min(60, max(20, i )) for i in bmi] 
#norm_bmi = [(i - 20) / (65 - 60) + 0.01 for i in bmi]  # here is a error before 20200605
norm_bmi = [(i - 20) / (60 - 20) + 0.01 for i in bmi] 
# bmi: 20 60

quit_time = df['quit_time'].tolist()
quit_time = [min(30, max(0, i)) for i in quit_time]
norm_quit_time = [(i - 0) / (30 - 0) + 0.01 for i in quit_time]
#min(quit_time) = 0, max(quit_time) = 23


race = df['race'].tolist()
for i in range(len(race)):
    if race[i] == 4:
        race[i] = 3
    if race[i] == 5:
        race[i] = 2.5
    if race[i] >= 6:
        race[i] = 1.5
norm_race = [(i - 1) / (3-1) + 0.01 for i in race]
# race: white 1, black: 2, Asian and Indian, Akaskan: 3, Nativive Hawaiian, 2.5, other: 1.5 

pkyr = df['pkyr'].tolist()
pkyr = [min(200, max(0, i)) for i in pkyr]
norm_pkyr = [(i - 0) / (200) + 0.01 for i in pkyr]
# pkyr: 30.0 200.0

df['age'] = norm_age
df['education'] = norm_education
df['bmi'] = norm_bmi
df['quit_time'] = norm_quit_time
df['race'] = norm_race
df['pkyr'] = norm_pkyr
df.to_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/spore/norm/spore_with1yr_byscan.csv', index = False)

```

```python
# copy from compare/clinical_spore/mcl_data.ipynb
# normalize nodule, bloodmark, with log
df =  pd.read_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/mcl/IPN_biomarker_norm.csv')
Log_nodsize, Log_bmark = [], []
for i, item in df.iterrows():
    
    if item['Nodule.Size.MM'] == item['Nodule.Size.MM']:
        log_nsize = min(5, max(1, np.log(item['Nodule.Size.MM']))) 
        Log_nodsize.append( (log_nsize - 1) / 4 + 0.01)
    else:
        Log_nodsize.append('')
    
    if item['CSI..CYFRA21.1..ng.mL.'] == item['CSI..CYFRA21.1..ng.mL.']:
        log_bmark = min(4, max(-4, np.log(item['CSI..CYFRA21.1..ng.mL.']))) 
        Log_bmark.append((log_bmark + 4) / (4 + 4) + 0.01)
    else:
        Log_bmark.append('')
        
df['Log_bmark'] = Log_bmark
df['Log_nodsize'] = Log_nodsize
df.to_csv('/nfs/masi/gaor2/data/Cotrain/clinical_spore/mcl/IPN_biomarker_norm.csv', index = False)
```
