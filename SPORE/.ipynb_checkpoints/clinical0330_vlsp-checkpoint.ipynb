```python
import matplotlib.pyplot as plt

labels = ['age',  'sex', 'race', 'LungRADS', 'height', 'weight', 'smok', 
          'packyear', 'since_quit', 'copd', 'education', 'Coronary Artery Calc', 'Qualify USPSTF', 'personal_cancer_history',
          'family_cancer_history']
weight = [ 0,0,0,0,0,0,0,0,0, 18.0, 18.0, 0, 0, 28., 64.]
print (len(labels), len(weight))
#weight = [round(a, 3) * int(100) for a in weight]
#weight = [21.8, 0.1, 0.1, 44.5, 54.3, 47.3, 18.2, 48.1, 63.7, 40.2, 51.4, 60.1, 27.7, 87.3, 89.0, 90.0, 91.0, 62.2]

x = range(len(labels))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots(figsize = (20, 5))
rects1 = ax.bar(x , weight, width, label='VLSP')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('missing rate (%)', fontsize = 15)
#ax.set_title('The importance of factors when caculating the risk')
ax.set_xticks(x)
ax.set_xticklabels(labels)
#ax.legend()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(True)
ax.spines["left"].set_visible(True)

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
plt.xticks(rotation = 45)
plt.yticks(fontsize = 15)
#plt.xticks(rotation = 45)
ax.set_xticklabels(labels, fontsize = 15)
ax.legend(fontsize = 15)
plt.show()
```

```python
df = pd.read_excel('/nfs/masi/SPORE/file/clinical/SPORE_nodules_7_20_anon.xlsx')

Sess = []
for i, item in df.iterrows():
    #print (item['DOS'])
    timestamp = str(item['DOS']).split(' ')[0].replace('-', '')
    ID = item['PID'].replace('SPORE_', '')
    Sess.append(ID + 'time' + timestamp)
df['sess'] = Sess
df.to_csv('/nfs/masi/SPORE/file/clinical/SPORE_nodules_7_20_anon.csv', index = False)
```

```python
df = pd.read_csv('/nfs/masi/SPORE/file/clinical/self_creat/Limitedhistory20200420.csv')

sess_fhist, sess_phist = {}, {}

for i, item in df.iterrows():
    if item['pid'] == item['pid']:
        sess_fhist[item['pid']] = item['fhist']
        sess_phist[item['pid']] = item['phist']

df = pd.read_csv('/nfs/masi/SPORE/file/clinical/self_creat/SPORE_nodules_7_20_anon.csv')
fhist, phist = [], [] 
for i, item in df.iterrows():
    try:
        fhist.append(sess_fhist[item['PID']])
        phist.append(sess_phist[item['PID']])

        
    except:
        
        fhist.append('')
        phist.append('')

df['fhist'] = fhist
df['phist'] = phist

df.to_csv('/nfs/masi/SPORE/file/clinical/self_creat/SPORE_nodules_7_20_anon.csv', index = False)
```

```python
df = pd.read_csv('/nfs/masi/SPORE/file/clinical/self_creat/Limitedhistory20200420.csv')

sess_age, sess_copd, sess_sex = {}, {}, {}

for i, item in df.iterrows():
    
    sess = item['sess']
    sess_age[sess] = item['age']
    sess_copd[sess] = item['copd']

df = pd.read_csv('/nfs/masi/SPORE/file/clinical/self_creat/SPORE_nodules_7_20_anon.csv')
age, copd, sex = [], [], []
for i, item in df.iterrows():
    try:
        age.append(sess_age[item['sess']])
        copd.append(sess_copd[item['sess']])

        
    except:
        print (item['sess'])
        age.append('')
        copd.append('')

df['age'] = age
df['copd'] = copd

df.to_csv('/nfs/masi/SPORE/file/clinical/self_creat/SPORE_nodules_7_20_anon.csv', index = False)
```

```python
df = pd.read_excel('/nfs/masi/SPORE/file/clinical/Limitedhistory4-20-20-lookup2019JunAge.xlsx')
ID_sex = {}
for i, item in df.iterrows():
    if item['SPORE'] == item['SPORE']:
        ID_sex[item['SPORE']] = item['sex']
        
df = pd.read_csv('/nfs/masi/SPORE/file/clinical/self_creat/SPORE_nodules_7_20_anon.csv')

sex =  []
for i, item in df.iterrows():
    try:
        sex.append(ID_sex[item['PID']])
    except:
        sex.append('')

df['sex'] = sex

df.to_csv('/nfs/masi/SPORE/file/clinical/self_creat/SPORE_nodules_7_20_anon.csv', index = False)
```

```python
df = pd.read_csv('/nfs/masi/SPORE/file/clinical/self_creat/SPORE_nodules_7_20_anon.csv')
uplobe, spic, nod_cnt,  = [], [], []

for i, item in df.iterrows():
    if item['Location'] != item['Location']:
        uplobe.append('')
    elif 'U' in item['Location']:
        uplobe.append(1)
    else:
        uplobe.append(0)
    
    if item['Spiculation (Yes/No)'] != item['Spiculation (Yes/No)']:
        spic.append('')
    elif 'No' in item['Spiculation (Yes/No)']:
        spic.append(0)
    else:
        spic.append(1)
    
    nod_cnt.append(1)
    
df['uplobe'] = uplobe
df['spic'] = spic
df['nod_cnt'] = nod_cnt

df.to_csv('/nfs/masi/SPORE/file/clinical/self_creat/SPORE_nodules_7_20_anon.csv', index = False)

    
```
